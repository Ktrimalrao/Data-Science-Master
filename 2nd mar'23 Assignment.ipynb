```python
"""Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
        Matplotlib."""

Ans.
    Matplotlib is a data visualization library in Python that allows you to create a wide range of visualizations to represent
    data in a graphical format. It provides a flexible & powerful framework for generating static, animated, interactive plots.
    
   
 * Line plots: Used to display the relationship between two continuous variables over a continuous interval.
                It is commonly used to show trends and patterns in data.

 * Scatter plots: Used to display the relationship between two continuous variables. 
            Each data point is represented as a dot on the plot,the position of the dot represents the values of the variables.

 * Bar charts: Used to compare categorical data or discrete variables. 
            It represents the data using rectangular bars, where the height or length of the bar represents the value of the variable.

 * Histograms: Used to represent the distribution of a continuous variable. 
            It divides the data into bins and shows the frequency or proportion of data points falling into each bin.

 * Pie charts: Used to represent the proportion or percentage of different categories in a dataset. 
                The size of each slice of the pie represents the relative value or proportion.

 * 3D plots: Used to visualize data in three-dimensional space. 
            It is useful for representing relationships between three continuous variables.
```

```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
        plot a scatter plot."""

# Note: Also add title, xlabel, and ylabel to the plot.
Ans.
    A scatter plot is a type of plot or graph that displays the relationship between two numerical variables.
    It uses Cartesian coordinates to represent the values of the variables as a collection of points on the plot.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
sizes = 300 * np.random.rand(50)
plt.scatter(x,y,c='#7B241C',s=sizes,alpha=0.894,marker='*')
plt.show()
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
        Use the following data:"""

Ans.
    The subplot() function in Matplotlibs Pyplot module is used to create a grid of subplots within a single figure. 
    It allows you to arrange multiple plots or axes in a grid-like structure, enabling you to display multiple 
    plots side by side or in a grid layout.
```

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# subplot(row,col,position)
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot 4')

plt.show()
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit,)
plt.title('DATA')
plt.xlabel('---- Company ----->')
plt.ylabel('---- Profit ----->')
```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot."""

Ans.
    A box plot, also known as a box-and-whisker plot, is a type of statistical visualization that provides a concise summary
    of the distribution of a dataset. It displays key statistical measures such as the median, quartiles, and potential outliers.
    
    
    Summarizing data distribution
    Comparing distributions
    
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data)
plt.xlabel('Box1')
plt.ylabel('Box2')
plt.title('* Box Plot *')
plt.show()
```
