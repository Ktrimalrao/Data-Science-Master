```python
"""Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot."""

Ans.
    Scatter Plot: scatter plot shows the relationship between two continuous variables. 
                    It is useful for visualizing patterns or trends in the data points.
        
    Relplot: The relplot() function in Seaborn allows you to create different types of relational plots,
            such as scatter plots, line plots, and categorical plots. 
            It provides a flexible interface to visualize the relationship between variables across multiple subsets of the dataset.
    
    CatPlot: The catplot() function is used to create categorical plots in Seaborn.
                It can create various types of plots like box plots, violin plots, bar plots, etc.
                This function is useful for comparing categories and understanding the distribution of categorical variables.
            
    JointPlot: Seaborn jointplot() function combines two different plots into one. 
                It shows the relationship between two variables using a scatter plot, while also providing individual
                univariate distributions for each variable using histograms or kernel density estimation.
            
    PairPlot: A pair plot, created using the pairplot() function in Seaborn, is used to visualize pairwise relationships
                between multiple variables in a dataset. It creates scatter plots for each combination of variables 
                and displays the correlation or distribution between them.
                Pair plots are helpful for identifying patterns and dependencies between variables.
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x=tips.total_bill , y = tips.tip)
```

```python
sns.relplot(x = tips.total_bill, y= tips.tip ,data = tips,hue='sex',style = 'size')
```

```python
tips['sex'].value_counts()
```

```python
sns.catplot(x='day' , y= 'smoker',data = tips,hue='sex')
```

```python
sns.jointplot(x=tips.total_bill,y = tips.smoker)
```

```python
sns.pairplot(tips,markers='^')
```

```python
"""Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
        "timepoint" and y = "signal" for different events and regions."""

sub = sns.load_dataset('fmri')
sub
```
