#### Q1. What is the mathematical formula for a linear SVM?

#### Ans.
The mathematical formula for a linear Support Vector Machine (SVM) can be described as follows:

Given a training dataset with features X and corresponding labels y, where X is an n x m matrix (n represents the number of data points, and m represents the number of features), and y is a vector of labels (-1 or 1), the objective of SVM is to find the optimal weight vector w and bias term b that define the hyperplane separating the classes.

The decision function for a linear SVM can be written as:

   ***f(x) = sign(w^T x + b)***

Here, x is a feature vector, w is the weight vector, b is the bias term, and sign() is the sign function that returns -1 if the argument is negative, 0 if it is zero, and 1 if it is positive.

The goal of SVM is to find the optimal w and b that minimize the classification error while maximizing the margin between the classes. This can be formulated as an optimization problem with the following constraints:

   ***minimize: (1/2) * ||w||^2***

subject to: y_i * (w^T x_i + b) >= 1 for all training examples (x_i, y_i)

The objective function aims to minimize the norm of the weight vector, which indirectly maximizes the margin between the classes. The constraints enforce that the decision function correctly classifies all training examples, ensuring that each data point lies on the correct side of the decision boundary.

The optimization problem is typically solved using quadratic programming techniques or convex optimization algorithms to find the optimal values of w and b that satisfy the constraints and minimize the objective function.

#### Q2. What is the objective function of a linear SVM?

#### Ans.
The objective function of a linear Support Vector Machine (SVM) is to find the optimal hyperplane that separates the classes while maximizing the margin between them. In the case of a linear SVM, the objective function can be formulated as follows:

   ***minimize: (1/2) * ||w||^2***

subject to: y_i * (w^T x_i + b) >= 1 for all training examples (x_i, y_i)

In this formulation, ||w||^2 represents the squared Euclidean norm of the weight vector w, which indirectly measures the margin between the classes. Minimizing ||w||^2 corresponds to maximizing the margin.

The second part of the objective function includes the constraints. For each training example (x_i, y_i), the expression y_i * (w^T x_i + b) represents the signed distance from the data point x_i to the decision hyperplane. The constraint y_i * (w^T x_i + b) >= 1 ensures that the decision function correctly classifies each training example, i.e., each data point lies on the correct side of the decision boundary.

The objective function of the linear SVM can be seen as a trade-off between maximizing the margin and minimizing the classification error. The SVM aims to find the values of w and b that satisfy the constraints while minimizing the norm of the weight vector, thus achieving the best separation between classes with the maximum margin.

#### Q3. What is the kernel trick in SVM?

#### Ans.
The kernel trick is a technique used in Support Vector Machines (SVM) to extend their capabilities to handle nonlinear decision boundaries without explicitly mapping the data into a higher-dimensional feature space. It allows SVM to work efficiently in the original input space while benefiting from the mathematical properties of a higher-dimensional space.

In traditional SVM, the decision boundary is a hyperplane that separates the classes in the feature space. However, when the data is not linearly separable, finding a linear decision boundary may not be effective. The kernel trick addresses this limitation by introducing a kernel function that implicitly maps the data into a higher-dimensional space where linear separation becomes possible.

Instead of explicitly computing the mapping of the data into a higher-dimensional space, the kernel function calculates the dot products between pairs of data points in the original space. The kernel function takes the form of K(x, x') = φ(x)·φ(x'), where φ(x) represents the transformation of a data point x into a higher-dimensional space.

By using the kernel trick, SVM can perform calculations in the original input space while implicitly operating in a higher-dimensional feature space. This avoids the computational cost and complexity of explicitly mapping the data into the higher-dimensional space.

Some commonly used kernel functions in SVM include:

1. Linear Kernel: K(x, x') = x·x'
   - This is equivalent to the dot product between the input feature vectors.

2. Polynomial Kernel: K(x, x') = (γ·(x·x') + r)^d
   - It computes the dot product raised to a power d, with an optional constant term r and a scaling factor γ.

3. Radial Basis Function (RBF) Kernel: K(x, x') = exp(-γ·||x - x'||^2)
   - It calculates the exponential of the squared Euclidean distance between the input feature vectors, scaled by the parameter γ.

By choosing an appropriate kernel function, SVM can effectively handle complex nonlinear decision boundaries in the original input space, making it a powerful and versatile machine learning algorithm.

#### Q4. What is the role of support vectors in SVM Explain with example

#### Ans.
In Support Vector Machines (SVM), support vectors play a crucial role in defining the decision boundary and determining the optimal hyperplane. Support vectors are the data points from the training dataset that lie closest to the decision boundary, representing the critical instances that influence the placement of the decision boundary.

The decision boundary in SVM is defined by a hyperplane, and the goal is to find the hyperplane that maximizes the margin between the classes while correctly classifying the data points. The support vectors are the data points that lie on or closest to the margin.

Support vectors are important because they determine the position and orientation of the decision boundary. The decision boundary is determined by the support vectors because they have the most influence on the margin. Removing or altering any other data point that is not a support vector will not affect the position of the decision boundary.

Let's consider a simple example to understand the role of support vectors in SVM:

Suppose we have a 2D dataset with two classes, labeled as red circles and blue triangles. The goal is to find a decision boundary that separates these two classes.

In this example, we have six data points. After training the SVM, let's assume that three of these data points become the support vectors. These support vectors are the closest points to the decision boundary.

The decision boundary of the SVM will be determined by these support vectors and will be positioned to maximize the margin between the classes while correctly classifying the data points. Any changes to the position or orientation of these support vectors will directly affect the decision boundary.

Other data points that are not support vectors do not contribute to the decision boundary in SVM. They might be farther from the decision boundary or lie within the margin region. These non-support vectors can be seen as "less important" for determining the decision boundary.

In summary, support vectors are the critical data points that define the decision boundary in SVM. They play a significant role in determining the optimal hyperplane and maximizing the margin between the classes.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Load IRIS dataset from sklearn

```python
from sklearn.datasets import load_iris
dataset = load_iris()
```

#### independent and dependent

```python
X = pd.DataFrame(dataset.data,columns=dataset.feature_names)
y = pd.DataFrame(dataset.target ,columns=['target'])
```

```python
X.head(3)
```

```python
y.tail(2)
```

#### train the dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.35)
```

####  Train a linear SVM classifier on the training set

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(X_train,y_train)
```

#### Predict the test dataset

```python
y_pred = svc.predict(X_test)
y_pred
```

#### accuracy of the model on the testing set

```python
from sklearn.metrics import accuracy_score
print('Accuracy Sscore :',accuracy_score(y_test,y_pred))
```

#### plot the dataset

```python
traing=pd.concat([X_train,y_train],axis=1)
sns.pairplot(traing,hue='target',markers=["o", "*", "v"])
plt.show()
```

#### Try different values of the regularisation parameter C and see how it affects the performance of the model.

```python
import warnings
warnings.filterwarnings('ignore')

C = np.linspace(0.1,10,10)
from sklearn.metrics import classification_report
accuracy = []
for i in C:
    svc = SVC(kernel='linear',C=i)
    svc.fit(X_train,y_train)
    y_pred = svc.predict(X_test)
    acc_score = accuracy_score(y_test,y_pred)
    accuracy.append(acc_score)
    print('C value:',i)
    print(classification_report(y_test,y_pred))
    print('----------------------------------------------------------------')
```

```python
import matplotlib.pyplot as plt
plt.plot(C, accuracy)
plt.xlabel('C value')
plt.ylabel('Testing Accuracy')
plt.title('Accuracies for different values of C')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test,y_pred)
sns.heatmap(cf,annot=True,cmap="crest")
```
